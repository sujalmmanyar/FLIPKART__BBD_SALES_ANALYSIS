```python
df['customer_name']=df['customer_name'].fillna('unknown')
```

```python
df[df['city'].isna()==1]
```

```python
df['city']=df['city'].fillna('NotListed')
```

```python
df['product_name'].unique()
```

```python
df['inferred_brand'] = df['product_name'].str.strip().str.split().str[0]
```

```python
df['brand'] = df['brand'].fillna(df['inferred_brand'])
```

```python
df.drop(columns=['inferred_brand'], inplace=True)
```

```python
df[df['category'].isna()==1]
```

```python
df['fake_cat'] = df['product_name'].str.strip().str.split().str[1]
```

```python
df['category'] = df['category'].fillna(df['fake_cat'])
```

```python
df['category'] = df['category'].replace({'Home':'Home Appliances'})
```

```python
df.drop(columns=['fake_cat'], inplace=True)
```

## FILLING NULL VALUES IN  COST_PRICE  SELLING_PRICE  GROUPED MEDIAN IMPUTATION ##


```python
df['cost_price_imputed'] = df['cost_price'].isna().astype(int)
df['selling_price_imputed'] = df['selling_price'].isna().astype(int)
```

```python
df['cost_price'] = pd.to_numeric(df['cost_price'], errors='coerce')
df['selling_price'] = pd.to_numeric(df['selling_price'], errors='coerce')
```

```python
category_cost_medians = df.groupby('category')['cost_price'].median()
category_selling_medians = df.groupby('category')['selling_price'].median()
```

```python
df['category_cost_median'] = df['category'].map(category_cost_medians)
df['category_selling_median'] = df['category'].map(category_selling_medians)
```

```python
df['cost_price'] = df['cost_price'].fillna(df['category_cost_median'])
df['selling_price'] = df['selling_price'].fillna(df['category_selling_median'])
```

```python
df.drop(['category_cost_median', 'category_selling_median'], axis=1, inplace=True)
```

## FINDING DUPLICATES ##

```python
df.duplicated().any()
```

```python
df[df.duplicated()]
```

```python
df=df.drop_duplicates()
```

## LETS RECALCULATE THE R.C.P.L.D.  ##


```python
df['item_revenue'] = df['selling_price'] * df['quantity']
```

```python
df['item_cost_total'] = df['item_cost'] * df['quantity']
```

```python
df['loss'] = df['item_cost_total'] - df['item_revenue']
df['loss'] = df['loss'].where(df['loss'] > 0, 0)
```

```python
df.drop('item_cost_total', axis=1, inplace=True)
```

```python
df['item_revenue'] = df['selling_price'] * df['quantity']
order_revenue = df.groupby('order_id')['item_revenue'].transform('sum')
df['revenue'] = order_revenue
```

```python
df['item_cost_total'] = df['item_cost'] * df['quantity']
order_cost = df.groupby('order_id')['item_cost_total'].transform('sum')
df['cost'] = order_cost
```

```python
df['profit'] = df['revenue'] - df['cost']
```
